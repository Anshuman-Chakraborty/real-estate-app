```python
# columns to drop -> link, property_id
df.drop(columns=['link','property_id'],inplace = True)
```

```python
# renaming the columns
df.rename(columns={'area':'price_per_sq_ft'},inplace = True)
df.head()
```

```python
df['society'].value_counts()
```

```python
df['society'].value_counts().shape
```

```python
import re
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★','',str(name)).strip()).str.lower()
```

```python
df['society'].value_counts().shape
```

```python
df['society'].value_counts()
```

```python
df['price'].value_counts()
```

```python
df[df['price']=='Price on Request']
```

```python
df = df[df['price']!='Price on Request']
```

```python
df.head()
```

```python
def normalized_prices(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
df.loc[:, 'price'] = df['price'].str.split(' ').apply(normalized_prices)
```

```python
df.head()
```

```python
# price per sq ft
df['price_per_sq_ft'].value_counts()
```

```python
df.loc[:,'price_per_sq_ft'] = df['price_per_sq_ft'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.strip().astype('float')
```

```python
df.head()
```

```python
# bedrooms
df['bedRoom'].value_counts()
```

```python
df['bedRoom'].isnull().sum()
```

```python
df[df['bedRoom'].isnull()]
```

```python
df = df[~df['bedRoom'].isnull()]
```

```python
df.loc[:,'bedRoom'] = df['bedRoom'].str.split(' ').str.get(0).astype('int')
```

```python
df.head()
```

```python
# bathrooms
df['bathroom'].value_counts()
```

```python
df['bathroom'].isnull().sum()
```

```python
df.loc[:,'bathroom'] = df['bathroom'].str.split().str.get(0).astype('int')
```

```python
df.head()
```

```python
# balcony
df['balcony'].value_counts()
```

```python
df['balcony'].isnull().sum()
```

```python
df.loc[:,'balcony'] = df['balcony'].str.split(' ').str.get(0).str.replace('No','0')
```

```python
df.head()
```

```python
# additionalRoom
df['additionalRoom'].value_counts()
```

```python
df['additionalRoom'].value_counts().shape
```

```python
df['additionalRoom'].isnull().sum()
```

```python
df['additionalRoom'].fillna('not available',inplace=True)
```

```python
df.loc[:,'additionalRoom'] = df['additionalRoom'].str.lower()
```

```python
df.head()
```

```python
# floor num
df['floorNum'].value_counts()
```

```python
df['floorNum'].isnull().sum()
```

```python
df[df['floorNum'].isnull()]
```

```python
df['floorNum'] = df['floorNum'].str.split(' ').str.get(0).replace('Ground','0').str.replace('Basement','-1').str.replace('Lower','0').str.extract(r'(\d+)')
```

```python
df.head()
```

```python
# facing
df['facing'].value_counts()
```

```python
df['facing'].isnull().sum()
```

```python
df['facing'].fillna('NA',inplace=True)
```

```python
df['facing'].value_counts()
```

```python
df.insert(loc=4,column='area',value=round((df['price']*10000000)/df['price_per_sq_ft']))
```
